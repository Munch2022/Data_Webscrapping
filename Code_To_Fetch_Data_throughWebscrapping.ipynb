```python
headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}

webpage= requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers). text

# webpge=requests.get(url, headers=headers).text    ; easy way of understanding and writing

# the o/p will be the same html code for that page in the website. to crosscheck u can right click on the 
# web page and goto 'view page source' ; you will see the same code. 
#  we will store the o/p to a variable coz we need to use it further
```

# till now we hit the url and got the html here. now we need to parse and get the necessasary information and thats teh main goal .

```python
# we will give this webpage to BeautifulSoup 

soup= BeautifulSoup(webpage, 'lxml')            

# with webpage we have to give a xml parser; 'lxml' is built in which helps in parsing teh html code
```

```python
# prettify will formate the html code in anicer way

print(soup.prettify()) 


# doing this you'll understand the content of the webpage and know how its organised
```

```python
# from teh webpage if u want to extract peice of info; like i want the one with h1 tage(h1<>)
# i should be getting a list of the elements that has h1 tage

soup.find_all('h1')       # may be h1 tag has only one item ; 
```

```python
#  now if i want just the 'List of comapnies in India' then how to extract

soup.find_all('h1')[0].text
```

```python
# lets say i want all the one tagged with h2

soup.find_all('h2')      # in h2 i got so many items where all the company names are mentioned
```

```python
# lets find how may are in the h2 list

len(soup.find_all('h2') )             # so it has 30, infact every page also had 30 comapnies listed
```

```python
# every company section has rating, reviews, no.of emps, how old is the company, etc
# if we want to extract multiple things from teh same tag, then we have to use the attribute along with tag
#  to know the line which is reposnible for that info just right click<inspect; hover the mose over teh section 
# and it'll hightlight that line. from there u can see and tags and attribute too... 

# if i want the rating, i see it belongs to p tag with attribute class= rating
soup.find_all('p', class_='rating')[:5]
```

```python
len(soup.find_all('p', class_='rating'))    # so for all 30 companies we have 30 ratings for each of them
```

```python
#  similarly we can extract reviews also. Its in a tag under attribute review count

soup.find_all('a', class_= 'review-count')[:5]

```

```python
len(soup.find_all('a', class_= 'review-count'))              # so gotit
```

```python
# it becomes difficult to hover on evry element and extract info. 
# so we will take the entire container of the company(say tcs, accenture, etc... ) and move ahead
# hover the mouse on the container of tcs itself. its teh line of code content wrapper in html

comp= soup.find_all('div', class_= 'company-content-wrapper')
len(comp)
```

```python
#  now lets run a loop and enter into every conatiner and extract the info(data) we want
# u can use find for single item in a list; findall for multiple 

name= []
rating= []
reviews= []
ctype= []
hq= []
old= []
employees= []
# desc= []

for i in comp:
    name.append(i.find('h2').text.strip())           # here im appending the name with o/p data
    rating.append(i.find_all('p', class_= 'rating')[0].text.strip())       # same as done above; just keeping in a loop 
    # above line u can use find so that one dont have to use 0th index for extraction
    # rating.append(i.find('p', class_= 'rating').text.strip())
    reviews.append(i.find('a', class_= 'review-count').text.strip())
    ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
    hq.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
    old.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
    employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
#     desc.append(comp[i].find_all('p', class_= 'description body-small')[0].text.strip()
    
    
# now lets create a dataframe from this 
dict1= {'Name': name, 'Rating': rating, 'Reviews': reviews, 'Type': ctype, 'Head Quarter': hq, 'Old': old, 'Employees': employees}

df= pd.DataFrame(dict1)  
    
   
```

# we extracted data from one page so far. Now lets run the loop for all the pages in the website; ie 333 to extract from all pages. but i would extract data from 100 pages only as it takes time

```python
#  lets run a loop for all the pages now
# within the loop copy the same previous code, hit the url, get webpage, create soup,company, etc


final_df= pd.DataFrame()    # created an empty df so that we load data extracted from all pages

for k in range(1, 100):
    url= 'https://www.ambitionbox.com/list-of-companies?page={}'.format(k)
    headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage= requests.get(url, headers=headers).text
    
    soup= BeautifulSoup(webpage, 'lxml')                   
    comp= soup.find_all('div', class_= 'company-content-wrapper')
    
    name= []
    rating= []
    reviews= []
    ctype= []
    hq= []
    old= []
    employees= []
    
    
    for i in comp:
        name.append(i.find('h2').text.strip())           # here im appending the name with o/p data
        rating.append(i.find_all('p', class_= 'rating')[0].text.strip())       # same as done above; just keeping in a loop 
        # above line u can use find so that one dont have to use 0th index for extraction
        # rating.append(i.find('p', class_= 'rating').text.strip())
        reviews.append(i.find('a', class_= 'review-count').text.strip())
        
        # using try and except as there we ae getting error coz of prbably some null values in that feild
        try:
            ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        
        try:
            hq.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
            
        try:
            old.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
            
        try:
            employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
  

    # now lets create a dataframe from this 
    dict1= {'Name': name, 'Rating': rating, 'Reviews': reviews, 'Type': ctype, 'Head Quarter': hq, 'Old': old, 'Employees': employees}

    df= pd.DataFrame(dict1) 
    
    
    final_df= final_df.append(df, ignore_index= True)
    
    
    
```

```python
#  since some rows are missing lets try ; Try and Except method to extract all 7 feilds
#  this time from all 333 pages

final_df= pd.DataFrame()    # created an empty df so that we load data extracted from all pages

for z in range(1, 333):
    url= 'https://www.ambitionbox.com/list-of-companies?page={}'.format(z)
    headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage= requests.get(url, headers=headers).text
    
    soup= BeautifulSoup(webpage, 'lxml')                   
    comp= soup.find_all('div', class_= 'company-content-wrapper')
    
    name= []
    rating= []
    reviews= []
    ctype= []
    hq= []
    old= []
    employees= []
    
    
    for i in comp:
        try:
            name.append(i.find('h2').text.strip())          
        except:
            name.append(np.nan)
            
        try:
            rating.append(i.find_all('p', class_= 'rating')[0].text.strip()) 
        except:
            rating.append(np.nan)
            
        try:
            reviews.append(i.find('a', class_= 'review-count').text.strip())
        except:
            reviews.append(np.nan)
        
        try:
            ctype.append(i.find_all('p', class_= 'infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        
        try:
            hq.append(i.find_all('p', class_= 'infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
            
        try:
            old.append(i.find_all('p', class_= 'infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
            
        try:
            employees.append(i.find_all('p', class_= 'infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
  

    # now lets create a dataframe from this 
    dict1= {'Name': name, 'Rating': rating, 'Reviews': reviews, 'Type': ctype, 'Head Quarter': hq, 'Old': old, 'Employees': employees}

    df= pd.DataFrame(dict1) 
    
    
    final_df= final_df.append(df, ignore_index= True)
    
    
    
```
